# Incidents Table

```python
data = {
    'incident_id': df['incident_id'],
    'date': df['date'],
    'state': df['state'],
    'city_or_county': df['city_or_county'],
    'address': df['address'],
    'n_killed': df['n_killed'],
    'n_injured': df['n_injured'],
    'incident_url': df['incident_url'],
    'source_url': df['source_url'],
    'congressional_district': df['congressional_district'],
    'incident_characteristics': df['incident_characteristics'],
    'latitude': df['latitude'],
    'location_description': df['location_description'],
    'longitude': df['longitude'],
    'n_guns_involved': df['n_guns_involved'],
    'notes': df['notes'],
    'sources': df['sources'],
    'state_house_district': df['state_house_district'],
    'state_senate_district': df['state_senate_district']
}

incidents = pd.DataFrame(data)
```

# Gun Table

```python
df2 = df[df['n_guns_involved']>0].reset_index()
gun_id = []
incident_id = []
```

```python
for i in range(len(df2)):
    for j in range(int(df2.loc[i]['n_guns_involved'])):
        incident_id_i = str(df2.loc[i]['incident_id'])
        index_id = str(j)
        gun_id.append(incident_id_i + index_id)
        incident_id.append(incident_id_i)
```

```python
gun_stolen = [0] * len(gun_id)
gun_type = [0] * len(gun_id)
counter = 0
```

```python
for i in range(len(df2)):
    for j in range(int(df2.loc[i]['n_guns_involved'])):
        index_id = str(j)
            
        stolen_keys = list(json.loads(df2['gun_stolen'].loc[i].replace("\'", "\"")).keys())
        type_keys = list(json.loads(df2['gun_type'].loc[i].replace("\'", "\"")).keys())
        
        if index_id in stolen_keys:
            stolen = json.loads(df2['gun_stolen'].loc[i].replace("\'", "\""))[index_id]
            gun_stolen[counter] = stolen
            
        if index_id in type_keys:
            g_type = json.loads(df2['gun_type'].loc[i].replace("\'", "\""))[index_id]
            gun_type[counter] = g_type      
            
        counter += 1
```

```python
data = {
    'gun_id': gun_id,
    'incident_id': incident_id,
    'stolen': gun_stolen,
    'type': gun_type
}

guns = pd.DataFrame(data)
```

# Participant Table

```python
df3 = df[df['participant_type']!='{}'].reset_index()
participant_id = []
incident_id = []
```

```python
for i in range(len(df3)):
    for j in list(json.loads(df3['participant_type'].loc[i].replace("\'", "\"")).keys()):
        incident_id_i = str(df3.loc[i]['incident_id'])
        index_id = str(j)
        participant_id.append(incident_id_i + index_id)
        incident_id.append(incident_id_i)
```

```python
part_age = [0] * len(participant_id)
part_age_group = [0] * len(participant_id)
part_gender = [0] * len(participant_id)
part_relationship = [0] * len(participant_id)
part_status = [0] * len(participant_id)
part_type = [0] * len(participant_id)
counter = 0
```

```python
for i in range(len(df3)):
    for j in list(json.loads(df3['participant_type'].loc[i].replace("\'", "\"")).keys()):
        index_id = str(j)
            
        age_keys = list(json.loads(df3['participant_age'].loc[i].replace("\'", "\"")).keys())
        age_group_keys = list(json.loads(df3['participant_age_group'].loc[i].replace("\'", "\"")).keys())
        gender_keys = list(json.loads(df3['participant_gender'].loc[i].replace("\'", "\"")).keys())
        relationship_keys = list(json.loads(df3['participant_relationship'].loc[i].replace("\'", "\"")).keys())
        status_keys = list(json.loads(df3['participant_status'].loc[i].replace("\'", "\"")).keys())
        type_keys = list(json.loads(df3['participant_type'].loc[i].replace("\'", "\"")).keys())
        
        if index_id in age_keys:
            age = json.loads(df3['participant_age'].loc[i].replace("\'", "\""))[index_id]
            part_age[counter] = age
            
        if index_id in age_group_keys:
            age_group = json.loads(df3['participant_age_group'].loc[i].replace("\'", "\""))[index_id]
            part_age_group[counter] = age_group     
            
        if index_id in gender_keys:
            gender = json.loads(df3['participant_gender'].loc[i].replace("\'", "\""))[index_id]
            part_gender[counter] = gender   
            
        if index_id in relationship_keys:
            relationship = json.loads(df3['participant_relationship'].loc[i].replace("\'", "\""))[index_id]
            part_relationship[counter] = relationship
            
        if index_id in status_keys:
            status = json.loads(df3['participant_status'].loc[i].replace("\'", "\""))[index_id]
            part_status[counter] = status
            
        if index_id in type_keys:
            p_type = json.loads(df3['participant_type'].loc[i].replace("\'", "\""))[index_id]
            part_type[counter] = p_type 
        
        counter += 1
```

```python
data = {
    'participant_id': participant_id,
    'incident_id': incident_id,
    'age': part_age,
    'age_group': part_age_group,
    'gender': part_gender,
    'relationship': part_relationship,
    'status': part_status,
    'type': part_type
}

participants = pd.DataFrame(data)
```
